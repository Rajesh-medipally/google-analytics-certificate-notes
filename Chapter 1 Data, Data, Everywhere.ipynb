## Big data analytics life cycle
Authors Thomas Erl, Wajid Khattak, and Paul Buhler proposed a big data analytics life cycle in their book, Big Data Fundamentals: Concepts, Drivers & Techniques. Their life cycle suggests phases divided into nine steps:

Business case evaluation
Data identification
Data acquisition and filtering
Data extraction
Data validation and cleaning 
Data aggregation and representation
Data analysis
Data visualization
Utilization of analysis results

# Data Strategy
Data strategy is the management of the people, processes, and tools used in data analysis. Let's break that down. You manage people by making sure they know how to use the right data to find solutions to the problem you're working on. For processes, it's about making sure the path to that solution is clear and accessible. For tools, you make sure the right technology is being used for the job

to understand easily, lets go with an example, consider lawn mowner. 
before mowning, first we checks the user manual to make sure how many are included and what are the safety measures to take, like glasses and shoes.
secound step, checks the fuel tank to make sure it is full. 
third step, looking all around the lawn to remove rocks bottles or something that troubles the lawn mover.

this is called strategy. lawn maybe driven by us or others in company but strategy is a process of making sure the process goes smoothly.
