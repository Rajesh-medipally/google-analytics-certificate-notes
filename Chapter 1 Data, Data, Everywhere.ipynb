## Big data analytics life cycle

Data identification <br>
Data acquisition and filtering <br>
Data extraction <br>
Data validation and cleaning <br> 
Data aggregation and representation <br>
Data analysis <br>

# Data Strategy
Data strategy is the management of the people, processes, and tools used in data analysis. Let's break that down. You manage people by making sure they know how to use the right data to find solutions to the problem you're working on. For processes, it's about making sure the path to that solution is clear and accessible. For tools, you make sure the right technology is being used for the job

<br>
<br>

Overall

Plan: Decide what kind of data is needed, how it will be managed, and who will be responsible for it.<br>
Capture: Collect or bring in data from a variety of different sources.<br>
Manage: Care for and maintain the data. This includes determining how and where it is stored and the tools used to do so.<br>
Analyze: Use the data to solve problems, make decisions, and support business goals.<br>
Archive: Keep relevant data stored for long-term and future reference.<br>
Destroy: Remove data from storage and delete any shared copies of the data.<br>#
