## How the data will be collected
Decide if you will collect the data using your own resources or receive (and possibly purchase it) from another party. Data that you collect yourself is called first-party data. <br>
## Data sources
If you don’t collect the data using your own resources, you might get data from second-party or third-party data providers. Second-party data is collected directly by another group and then sold. Third-party data is sold by a provider that didn’t collect the data themselves. Third-party data might come from a number of different sources. <br>
## How much data to collect
If you are collecting your own data, make reasonable decisions about sample size. A random sample from existing data might be fine for some projects. Other projects might need more strategic data collection to focus on certain criteria. Each project has its own needs. <br>
## Time frame
If you are collecting your own data, decide how long you will need to collect it, especially if you are tracking trends over a long period of time. If you need an immediate answer, you might not have time to collect new data. In this case, you would need to use historical data that already exists.

## internal, external data
internal,
Data that lives inside a company’s own systems. <br> ex, Wages of employees in an organisation. <br>
external,
Data that lives outside of a company.<br> ex, Credit reports for customers of an auto dealership.

## Continuous, Discrete
Continuous data,
Data that is measured and can have almost any numeric value.<br>
ex,Height of kids in third grade classes (52.5 inches, 65.7 inches)<br>
Discrete data,
Data that is counted and has a limited number of values.(it cannot have any after decimal value at most).<br>
ex, Number of people who visit a hospital on a daily basis (10, 20, 200).
## Qualitative, Quantitative
Qualitative,
Subjective and explanatory measures of qualities and characteristics.<br>
ex,Favorite brands of most loyal customers.<br>
Quantitative,
Specific and objective measures of numerical facts.<br>
ex,Percentage of board certified doctors who are women.
## Nominal, Ordinal 
Nominal,
A type of qualitative data that isn’t categorized with a set order.<br>
ex, First time customer, returning customer, regular customer or New job applicant, existing applicant, internal applicant
(this is basically means, without having perticular border or context data is generated)<br>
Ordinal,
 A type of qualitative data with a set order or scale.<br>
ex, Movie ratings (number of stars: 1 star, 2 stars, 3 stars)<br>
## Structured, Unstructured
Structured data,
Data organized in a certain format, like rows and columns.<br>
ex,Expense reports  or Tax returns<br>
Unstructured data,
Data that isn’t organized in any easily identifiable manner. <br>
ex,Social media posts.<br>



## data modeling
this is a process, where, to design a model for data. as a data analyst mostly you dont get this to do. but for understanding consider it as a blue print of a house. the electrician, plumber,builder need the blue print but each of there perspectives are totally different. this is as same as data model. 



1) Conceptual data modeling gives a high-level view of the data structure, such as how data interacts across an organization. For example, a conceptual data model may be used to define the business requirements for a new database. A conceptual data model doesn't contain technical details. 

2) Logical data modeling focuses on the technical details of a database such as relationships, attributes, and entities. For example, a logical data model defines how individual records are uniquely identified in a database. But it doesn't spell out actual names of database tables. That's the job of a physical data model.

3) Physical data modeling depicts how a database operates. A physical data model defines all entities and attributes used; for example, it includes table names, column names, and data types for the database.

## The open data
In data analytics, open data is part of data ethics, which has to do with using data ethically. Openness refers to free access, usage, and sharing of data.
<br>
But for data to be considered open, it has to:
<br>
Be available and accessible to the public as a complete dataset
<br>
Be provided under terms that allow it to be reused and redistributed
<br>
Allow universal participation so that anyone can use, reuse, and redistribute the data
<br>
Data can only be considered open when it meets all three of these standards. <br>

#### Meta data
this is data about data.<br>
Elements of metadata<br>
Title and description ->
What is the name of the file or website you are examining? What type of content does it contain?
<br>
Tags and categories->
What is the general overview of the data that you have? Is the data indexed or described in a specific way? 
<br>
Who created it and when->
Where did the data come from, and when was it created? Is it recent, or has it existed for a long time?
<br>
Who last modified it and when->
Were any changes made to the data?  If yes, were the modifications recent?
<br>
Who can access or update it->
Is this dataset public? Are special permissions needed to customize or modify the dataset?<br>

examples :-
Photos->
Whenever a photo is captured with a camera, metadata such as camera filename, date, time, and geolocation are gathered and saved with it.
<br>
Emails->
When an email is sent or received, there is lots of visible metadata such as subject line, the sender, the recipient and date and time sent. There is also hidden metadata that includes server names, IP addresses, HTML format, and software details.<br>

## Organizing and protecting data
#### Benifits fo Organizing your data.
makes it easier to find and use, <br>
helps you avoid making mistakes during your analysis and helps to protect it.

Best practices for keeping files organized<br>

Create folders and subfolders in a logical hierarchy so related files are stored together.
<br>
Separate ongoing from completed work so your current project files are easier to find. Archive older files in a separate folder, or in an external storage location.
<br>
If your files aren't automatically backed up, manually back them up often to avoid losing important work.
<br>


When creating a file structure and naming convention pattern for a project, you should always:<br>
<br><br>
Work out your conventions early in your project. The earlier you start, the more organized you’ll be. 
<br><br>
Align file naming conventions with your team. Conventions are most useful when everyone follows them.
<br><br>
Make sure filenames are meaningful. Stick to a consistent pattern that contains the most useful information needed.
<br><br>
Keep file names short and to the point.
