the one of the main points of being data analyst is to ask the right question. for that there are some paths can follow. <br>

for example, if some one says this sandwich is good isnt it? this question has chance of explanation. <br>
another one like, do you like growing here ? they can say yes or no . this also doesnt give you information. <br>

the strategy asking qeustions is <br>
S -> Specific <br>
M -> Measurable <br>
A -> Action-Oriented <br>
R -> Relevant <br>
T -> Time - Bound <br>


## Specific 

asking a focused question rather than general. <br>

are the kids getting excerice these days ? is a too normal question, where we may or may not can explain. <br>
asking like, the percentage of kids getting regular activities for an hour a day per 5 days a week. ? this gives details information of what we needed actually.

overall, try to focus on a single aspect at a time, rather than normalizing the situation.

## Measurable

measurable questions, that, those can be quantified. <br>

it means having a number that can give the brief answer. <br>

for example, asking how the video gone viral ? is too general.
how many times the video is been shared in the first week ? will give measurement to understand the viral percentage.

## Action-Oriented 

these questions help to change the situation.  <br>

like, asking for how to make our product recycle more ? <br> is normal
asking what design can bring more recycle ? <br> making it action oriented. where, you can see a way to solve it.

## Relevant

asking relevant question is really important to solve an issue. the question asked has to be very appropriate.

for example, consider a situation that one kind of frog species are disappiaring and needed to ask a right and relevent question. <br>

aksing like, how many are frog species are disappeared in an year ? can be not an accurate and irrilavent question. because, it is not focused on what happend and why. <br>

asking like, what are the environmental factors that effect the corolina from 2000 to 2004 that caused the frog species disappear.? this gives a focused and relevant option and it makes the range of possibilities less.
<br>

in this  questioning we included a specific place and time so that finding makes easy rather than asking overall.

## time bound 

the qeustion has to be time bounded, so that, can focus on relevant data with less range of possibilities and accuracy.
