## Analysis
The process used to make sense of data collected <br>
The goal of analysis is to identify trends and relationships within the data so that you can accurately answer the question you're asking. <br>
## The 4 Phases of Analysis
1. Organize data -> it is like a registary where data is organized ( contact list in phone)
2. Format and adjust data -> filtering and sorting data
3. Get input from others -> the updates about working data
4. Transform the data -> identifying relationships, formats and calculating results on the base of data you have 

## Organize data for analysis

Most of the data you'll use in your analysis will be organized in tables. Tables help you organize similar kinds of data into categories and subject areas that you can focus on as you analyze. 
<br>
organizing data means, ability to find what we wanted in a visual procedure.<br>
most of the organizing part includes sorting and filtering the data.<br>
Sorting is when you arrange data into a meaningful order to make it easier to understand, analyze, and visualize. <br>
Filtering is showing only the data that meets a specific criteria while hiding the rest.<br>
