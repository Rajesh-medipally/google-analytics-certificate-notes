##  Date format in google sheets
For its internal database, Google Sheets stores all dates as integer numbers. Not sequences of a day, month, and year as we got used to seeing, but simple integers:<br>
<br>
<br>1 for December 31, 1899<br>
2 for January 1, 1900<br>
102 for November 4, 1900 (100 days after January 1, 1900)<br>
and so on.<br>
Unlike Excel that cannot store dates as negative numbers, in Google, for dates prior to December 31, 1899, the numbers will be negative:<br>

-1 for December 29, 1899<br>
-2 for December 28, 1899<br>
-102 for September 19, 1899<br>
etc.<br>
Regardless of how Google Sheets formats dates for you to see in cells, spreadsheets always store them as integers. It's an automatic Google Sheets date format that helps to treat dates correctly.<br>


changing the date format in google sheet is by clicking the option in the menu bar format.

#### JOINS used in SQL
JOIN is SQL clause that is used to combined rows from two or more tables based on related columns.
<br>
there are 4 common joins<br>
inner<br>
left<br>
right<br>
outer<br>




An inner JOIN is a function that returns records with matching values in both tables. <br>
A LEFT JOIN is a function that will return all the records from the left table and only the matching records from the right table.<br>
RIGHT JOIN does the opposite. It will return all records from the right table and only the matching records from the left. <br>
OUTER join combines RIGHT and LEFT JOIN to return all matching records in both tables.



to understand the joins consider the tables below 'Orders' and 'Customers'


in these 2 tables common column is customer id by using that, we can combine the tables.
